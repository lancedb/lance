```python
import lance
import pyarrow as pa
```

```python

data = pa.table({
    "id" : range(10),
    
    "value": [True, False] * 5,
    
})

ds = lance.write_dataset(data, "memory://")
ds.delete("NOT value")

ds.to_table(filter="value")
```

```python
ds.create_scalar_index("value","ZONEMAP")
ds.to_table(filter="value")
```
